```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('demand_limpio.csv')
df.columns
df.head(30)
```

```python
# 2. Convertir los datos a los tipos necesarios
df['date'] = pd.to_datetime(df['date'])
df['order_Demand'] = pd.to_numeric(df['order_demand'], errors='coerce')
# Crear nuevas columnas para la semana, el mes y el año
df['week'] = df['date'].dt.isocalendar().week
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
```

```python
df['order_Demand'].drop(0, inplace=True)
df.columns
```

```python
l = ["002", "008", "0012", "001", "0033", "0036", "0038", "0071", "0083", "0093", "0094", "0100", "0116"]
df['product_code'] = df['product_code'].apply(lambda x: x.split('_')[1])
df['product_code'] = df['product_code'].apply(lambda x: x if x in l else None)
df = df.dropna()
df['product_code'].head(30)
df.columns
df['product_code'].unique()
```

```python
# Cambiar la columna 'date' a tipo datetime df
df['date'] = pd.to_datetime(df['date'])


# Agrupar por 'warehouse', 'product_code', 'year', 'month', 'week', y 'date', y sumar la 'order_demand'
grouped_df = df.groupby(['product_code', 'warehouse', 'year', 'month', 'week', 'date'])['order_demand'].sum().reset_index()

# Ordenar los resultados por 'date' dentro de cada 'warehouse'
grouped_df_sorted = grouped_df.sort_values(by=['product_code', 'date'])

# Crear DataFrames individuales para cada almacén
df_whse_j = grouped_df_sorted[grouped_df_sorted['warehouse'] == 'Whse_J']
df_whse_s = grouped_df_sorted[grouped_df_sorted['warehouse'] == 'Whse_S']
df_whse_c = grouped_df_sorted[grouped_df_sorted['warehouse'] == 'Whse_C']
df_whse_a = grouped_df_sorted[grouped_df_sorted['warehouse'] == 'Whse_A']
```

```python
def me(last_fx: float, last_mu: float, rho=0.8) -> float:
    return rho*last_mu + (1-rho)*last_fx
```

```python
grouped_whs_c[0].head(30)
```

```python
for df_product in grouped_whs_c:
    df_product['storage'] = 0
    df_product['mu'] = 0
    df_product['mu_a'] = 0
    df_product['pi'] = 0
    df_product['pir'] = 0
    
    df_product['storage'].iloc[0] = df_product['order_demand'].iloc[0]*1.3
    cont = 0    
    mu = df_product["order_demand"].mean()
    
    for od in df_product['order_demand']:
        
        df_product['mu'].iloc[cont] = mu
        mu = me(od , mu, rho =0.85)
        mu_a = me(mu, od, rho = 0.85)
        df_product['mu_a'].iloc[cont] = mu_a
        df_product['pi'].iloc[cont] = abs(df_product["storage"].iloc[cont] -mu - od)
        df_product['pir'].iloc[cont] = df_product['pi'].iloc[cont]
        
        try:
            df_product['storage'].iloc[cont+1] = df_product['storage'].iloc[cont] - df_product['order_demand'].iloc[cont] + df_product['pi'].iloc[cont]
        except Exception:
            pass
        
        cont +=1
        df_product.to_csv(f"whs_c_{df_product['product_code'].iloc[0]}.csv")
           
            
        
    
```
